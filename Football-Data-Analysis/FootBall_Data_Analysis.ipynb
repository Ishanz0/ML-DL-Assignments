```python
salary= dataset[['short_name', 'wage_eur']]
salary
```

```python
salary=salary.sort_values(by=['wage_eur'],ascending = False)
```

```python
plt.figure(figsize=(20,5))
plt.bar(list(salary['short_name'])[0:10],list(salary['wage_eur'])[0:10],color='green')
plt.show()
```
